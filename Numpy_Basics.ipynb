# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3])
```

```python
x
```

```python
plt.plot(x)
#plt.scatter(x)
```

```python
y=np.array(x+1)
```

```python
plt.plot(x,y)
plt.scatter(x,y,color="red")
plt.ylabel("y-axis")
plt.title("Simple plot")
plt.legend()

plt.show()
```
